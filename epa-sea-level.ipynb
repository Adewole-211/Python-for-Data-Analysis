```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv ("epa-sea-level.csv")
df = df.dropna(subset=["Year","CSIRO Adjusted Sea Level"])
y = df["CSIRO Adjusted Sea Level"]
x = df["Year"]
```

```python
fig, ax = plt.subplots()
plt.scatter(x,y)
```

```python
res1 = linregress(x, y)
x_pred1 = pd.Series(range(1880, 2051))
y_pred1 = res1.slope * x_pred1 + res1.intercept
```

```python
df_recent = df[df['Year'] >= 2000]
x2 = df_recent['Year']
y2 = df_recent['CSIRO Adjusted Sea Level']
res2 = linregress(x2, y2)
x_pred2 = pd.Series(range(2000, 2051))
y_pred2 = res2.slope * x_pred2 + res2.intercept
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y, alpha=0.6, label="Actual Data")
plt.plot(x_pred1, y_pred1, "r", label="Regression Line: 1880-2050")
plt.plot(x_pred2, y_pred2, "g", label="Regression Line: 2000-2050")
plt.xlabel("Year")
plt.ylabel("Sea Level(Inches)")
plt.title("Rise In Sea Level")
plt.legend()
plt.show()
```

```python
def draw_cat_plot(): #this starts a function that assumes you already have a data frame, in the context, we are talking about the med_df frame.
    #with  the columns, cholesterol, gluc, yadayadayada.
    # Melt dataframe
    df_cat = pd.melt( # then we reshape it from wide to long in with melt
        med_df,
        id_vars=['cardio'], # so cardio is an identifier that stays in its own column. most likely what the other columns revolve around. This then stacks,
        #the selected columns into 2 columns, the being that data inputs 1 and 0 for cardio. 
        value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active']
    )

    # Group & count
    df_cat = df_cat.groupby(['cardio', 'variable', 'value'], as_index=False).size() # groups it by "cardio" values 1 and 0. Additional the .size() part 
    #counts the number of rows in each group
    df_cat.rename(columns={'size': 'total'}, inplace=True) # this section essentially renames the columns from size to total. 

    # this section is where the action starts and we get to draw the graph. 
    g = sns.catplot(
        data=df_cat,
        x='variable', # this gives bars for each category 
        y='total', # this means the bar hieght equals the count that has been computed 
        hue='value', #this splits the bar by the 0/1 of that variable.
        col='cardio', #makes seperates panels (facets) for cardio=0 and cardio=1.
        kind='bar' # due to it being aggregated counts, we will use bar charts, not "Kind=count", which would count from the raw data, skipping all that we
        #we added.
    )

    # Get the figure and return it
    fig = g.fig
    return fig #this is the answer to number 8
```

```python
fig = draw_cat_plot()
```

```python
def draw_heat_map():
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np

    # Clean the data
    med_df_heat = med_df[
        (med_df['ap_lo'] <= med_df['ap_hi']) &
        (med_df['height'] >= med_df['height'].quantile(0.025)) &
        (med_df['height'] <= med_df['height'].quantile(0.975)) &
        (med_df['weight'] >= med_df['weight'].quantile(0.025)) &
        (med_df['weight'] <= med_df['weight'].quantile(0.975))
    ]

    # Compute correlation matrix
    corr = med_df_heat.corr(method="pearson")

    # Generate mask for upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Set up matplotlib figure
    fig, ax = plt.subplots(figsize=(12, 12))

    # Draw heatmap
    sns.heatmap(
        corr,
        mask=mask,
        annot=True,
        fmt=".1f",
        square=True,
        linewidths=1,
        center=0,
        cbar_kws={"shrink": 0.5},
        ax=ax
    )

    # Save to file
    fig.savefig("heatmap.png")

    # Show inline in Jupyter
    plt.show()

    # Return ax for further tweaking
    return ax
        

```

```python
draw_heat_map()
```
